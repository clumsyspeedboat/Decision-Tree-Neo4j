```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.tree import plot_tree
from sklearn.model_selection import cross_val_score
from timeit import default_timer as timer
import time
```

```python
path = r"C:\Users\49171\Desktop\Graph-Database-Learning-Algorithms-Neo4j\Dataset 3 - Flu Classification\Flu Classification.csv"
df = pd.read_csv(path)
df.head()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
x.head()
```

```python
# Cross Validation 10 folds : Gini
a = timer()
dt1 = DecisionTreeClassifier(criterion = "gini")
scores = cross_val_score(dt1, x, y, cv=10)
b = timer()

delta = b - a
print("Accuracy: " + str(scores.mean()) + " %")
print("Time taken to generate tree: " + str(delta) + " second")
```

```python
# Cross Validation 10 folds : Info Gain
a = timer()
dt1 = DecisionTreeClassifier(criterion = "entropy")
scores = cross_val_score(dt1, x, y, cv=10)
b = timer()

delta = b - a
print("Accuracy: " + str(scores.mean()) + " second")
print("Time taken to generate tree: " + str(delta) + " second")
```
